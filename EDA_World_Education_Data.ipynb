As there are 3665 unique indicators, most of which are null, we divide the task of selecting relavant indicators into two steps:

I. Determine the indicators which have an acceptable level of missing data.

II. choose indicators relaavant to this project out of these indicators determined in step I.



Defining a function to select indicators with an acceptable level of missing data:

````python

# Defining function to select the indicators for analysis

def select_parameters(series, total_len, tolerance):

  # Define function to check for missing values
  def check_missing_values(series):
      return series.isna().sum()

  # Applying the check missing values function
  missing = series.apply(check_missing_values).sort_values()


  # Filtering  to include only such indicators
  filtered = missing[missing < tolerance]

  return filtered
````

 #### Chart - 3 : Which countries have highest and lowest mean per-capita GDP and how did it vary from 1990 to 2005?

````python
# defining a function to select n top and bottom countries and plot their trends over the years for any given indicator

def top_bottom_countries(df, Indicator_Name, years, n):
    try:
        # Check if the DataFrame contains the necessary columns
        if 'Indicator Name' not in df.columns or any(year not in df.columns for year in years):
            raise ValueError("The DataFrame must contain the 'Indicator Name' column and the specified year columns.")

        # Make a DataFrame with only the relevant indicator
        df_filtered = df[df['Indicator Name'] == Indicator_Name]

        if df_filtered.empty:
            raise ValueError(f"No data found for the indicator '{Indicator_Name}'.")

        # Check if the years are valid
        if not all(year in df_filtered.columns for year in years):
            raise ValueError("Some of the specified years are not in the DataFrame columns.")

        # Identifying top and bottom countries
        top_countries = df_filtered[years].mean(axis=1).nlargest(n).index

        bottom_countries = df_filtered[years].mean(axis=1).nsmallest(n).index

        top_and_bottom_countries = top_countries.union(bottom_countries)

        # Making a DataFrame with top and bottom countries

        combined_df = df_filtered[df_filtered.index.isin(top_and_bottom_countries)][years]

        return top_countries, bottom_countries, combined_df

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except KeyError as ke:
        print(f"KeyError: {ke}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


````

````python
#  Plotting GDP trends for top 5 and bottom 5 countries over the years

# choosing the dataset and indicators

df = edstats_countries_past.copy()

df.set_index('Country Name', inplace=True)

Indicator_Name = 'GDP per capita, PPP (constant 2011 international $)'

years = ['1990', '1995', '2000', '2005']

n = 5

# Applying the above defined top_bottom_countries function

top_countries, bottom_countries, combined_df = top_bottom_countries(df, Indicator_Name, years, n )


# Plotting the trends
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

cmap = plt.get_cmap('tab20')

mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=[cmap(i) for i in np.linspace(0, 1, 10)])

# Plotting the trends of top countries

for country in top_countries:
  country_data = combined_df.loc[country].transpose()
  ax1.plot(country_data.index, country_data.values, label=country, linestyle='-',  marker='o' )

# Setting labels for the top countries plot
  ax1.set_xlabel('Year')
  ax1.set_ylabel(Indicator_Name)
  ax1.set_title('Trends of countries with \n highest GDP per capita (PPP)')
  ax1.legend()
  ax1.grid(True)

# Plotting the trends of bottom countries
for country in bottom_countries:
  country_data = combined_df.loc[country].transpose()
  ax2.plot(country_data.index, country_data.values, label=country, linestyle='--', linewidth=2, marker='o')

# Setting labels for the bottom countries plot
  ax2.set_xlabel('Year')
  ax2.set_ylabel(Indicator_Name)
  ax2.set_title('Trends of ountries with  \nlowest GDP per capita (PPP)')
  ax2.legend()
  ax2.grid(True)

plt.tight_layout()
plt.show()



````
